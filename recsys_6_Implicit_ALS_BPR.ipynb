## Детали и реализация алгоритма ALS.

ALS ( альтернирующие наименьшие квадраты) - это метод факторизации матриц пользователей и предметов; в ALS матрицы пользователей и предметов поочередно обновляются для аппроксимации матрицы предсказаний.

### Табличное выражение ALS

Основная идея ALS заключается в нахождении матрицы пользователей $\mathbf{U}$ и матрицы элементов $\mathbf{I}$. Матрица оценки $\mathbf{R}$ аппроксимируется следующим образом.

$$
\mathbf{R} \approx \mathbf{U} \mathbf{I}^T
$$

Здесь ALS решает следующую задачу минимизации.

$$
\min_{\mathbf{U}, \mathbf{I}} \|\| \mathbf{R} - \mathbf{U} \mathbf{I}^T \|\|^2_F + \lambda ( \|\| \mathbf{U} \|\|^2_F + \|\| \mathbf{I} \|\|^2_F )
$$

Где $\|\| \cdot \|\|_F$ представляет норму Фробениуса, а $\lambda$ - параметр регуляризации.

### Реализация ALS

Ниже приведен пример реализации ALS с использованием библиотеки IMPLICIT.


```python
import implicit
from scipy.sparse import coo_matrix
from pprint import pprint

# Загрузка и предварительная обработка данных
df = pd.read_csv("u.data", sep="\t", names=["user_id", "item_id", "rating", "timestamp"])
rows = df["user_id"].astype(int)
cols = df["item_id"].astype(int)
values = df["rating"].astype(float)

df.head()
```

```python
# Создание матрицы оценок
R = coo_matrix((values, (rows, cols)))

# Преобразование матрицы coo_matrix в матрицу csr_matrix
R = R.tocsr()

# Обучение модели ALS
model = implicit.als.AlternatingLeastSquares(factors=20, regularization=0.1, iterations=50)
model.fit(R)

# Матрицы пользователей и предметов
U = model.user_factors
I = model.item_factors

# Отображение результатов
pprint(U.round(2))
pprint(I.round(2))
```

## Детали и реализация алгоритма BPR

### Формулировка BPR
BPR максимизирует вероятность того, что пользователь предпочтет один товар другому. В частности, она максимизирует следующую функцию логарифмического правдоподобия:

$$
\sum_{(u,i,j) \in D} \ln \sigma (\hat{x}\_{u,i} - \hat{x}\_{u,j}) + \lambda \|\| \Theta \|\|^2
$$

где σ - сигмоидальная функция, x^u,i - оценка пользователя u для элемента i, D - набор данных, а Θ - параметры модели.

### Реализация BPR

```python
import implicit
from scipy.sparse import coo_matrix
from pprint import pprint

# Загрузка и предварительная обработка данных
df = pd.read_csv("u.data", sep="\t", names=["user_id", "item_id", "rating", "timestamp"])
rows = df["user_id"].astype(int)
cols = df["item_id"].astype(int)
values = df["rating"].astype(float)

df.head()
```

```python
# Создание матрицы оценок
R = coo_matrix((values, (rows, cols)))

# Преобразование матрицы coo_matrix в матрицу csr_matrix
R = R.tocsr()

# Обучение модели BPR
model = implicit.bpr.BayesianPersonalizedRanking(factors=20, regularization=0.1, iterations=50)
model.fit(R)

# Матрицы пользователя и элементов
U = model.user_factors
I = model.item_factors

# Вывод результатов на экран
pprint(U.round(2))
pprint(I.round(2))
```

## Пример реализации.

В качестве конкретного примера реализации модели ALS и BPR построены на основе набора данных movielens-100k.

### Подготовка данных

Сначала данные загружаются и проходят предварительную обработку.

```python
import implicit
import pandas as pd
import numpy as np

from scipy.sparse import coo_matrix
from sklearn.model_selection import train_test_split

# Загрузка и предварительная обработка данных
df = pd.read_csv("u.data", sep="\t", names=["user_id", "item_id", "rating", "timestamp"])

# Разделение на обучающие и тестовые данные
# Установите для параметра stratify значение True, чтобы распределение оставалось неизменным после разделения данных.
train, test = train_test_split(df, test_size=0.2, stratify=df["user_id"], shuffle=True, random_state=seed)

# Создание матрицы оценок
train_matrix = coo_matrix((train["rating"], (train["user_id"], train["item_id"])))
test_matrix = coo_matrix((test["rating"], (test["user_id"], test["item_id"])))

# Преобразование матрицы coo_matrix в матрицу csr_matrix
train_matrix = train_matrix.tocsr()
test_matrix = test_matrix.tocsr()
```

### Обучение и оценка моделей ALS

Следующий шаг - обучение и оценка модели ALS.

```python
def get_precision(true_matrix, pred_matrix, k=10):
    """
    Функция расчета точности

    Параметры:.
    - true_matrix (coo_matrix): матрица фактической оценки
    - pred_matrix (ndarray): матрица предсказанных оценок
    - k (int): количество элементов top_k при вычислении точности

    Возвращает:
    - precision (float): точность
    """
    # Преобразование матрицы фактических оценок в список
    true_items = true_matrix.tolil().rows

    # Получение индекса предсказанных элементов
    pred_items = np.argsort(-pred_matrix, axis=1)[:, :k]

    # Рассчитайте точность на одного пользователя.
    precisions = []
    for user_id in range(len(true_items)):
        true_set = set(true_items[user_id])
        pred_set = set(pred_items[user_id])

        if len(true_set) > 0:
            precision = len(true_set & pred_set) / min(len(true_set), k)
            precisions.append(precision)

    # Рассчитайте среднюю точность.
    return np.mean(precisions)
```

```python
# Обучение модели ALS
als_model = implicit.als.AlternatingLeastSquares(factors=20, regularization=0.1, iterations=50)
als_model.fit(train_matrix)

# Предсказания для тестовых данных
test_predictions = als_model.recommend_all(test_matrix)


# Примеры, подтверждающие правильность выбора
true_matrix = test_matrix  # Матрица фактической оценки тестовых данных
pred_matrix = als_model.recommend_all(test_matrix)  # Предсказание по модели ALS

precision = get_precision(true_matrix, pred_matrix)
print(f"ALS Model Precision: {precision:.3f}")
```

### Обучение и оценка моделей BPR

Аналогичным образом обучите и оцените модель BPR.

```python
# Обучение модели BPR
bpr_model = implicit.bpr.BayesianPersonalizedRanking(factors=20, regularization=0.1, iterations=50)
bpr_model.fit(train_matrix)

# Предсказания для тестовых данных
test_predictions = bpr_model.recommend_all(test_matrix)

# Оценка точности
precision = get_precision(test_matrix, test_predictions)
print(f"BPR-модель Precision : {precision:.3f}")
```

Загрузка и предварительная обработка данных

```python
events = pd.read_csv(path+"/events.csv")

# оставим только взаимодействия, которые можно считать implicit feedback (например, 'view')
events = events[events['event'] == 'view']

# закодируем id-шники
user_map = {id_: i for i, id_ in enumerate(events['visitorid'].unique())}
item_map = {id_: i for i, id_ in enumerate(events['itemid'].unique())}
events['user_id'] = events['visitorid'].map(user_map)
events['item_id'] = events['itemid'].map(item_map)

# аггрегируем количество просмотров (view count)
interaction_matrix = events.groupby(['user_id', 'item_id']).size().reset_index(name='strength')

# создание разреженной матрицы
rows = interaction_matrix['user_id'].values # индексы пользователей
cols = interaction_matrix['item_id'].values  # индексы товаров
data = interaction_matrix['strength'].values # сила взаимодействия

R = coo_matrix((data, (rows, cols))).tocsr()
R
```

```python
print(R)
```

```python
from sklearn.model_selection import train_test_split
from scipy.sparse import lil_matrix

# создаём список взаимодействий
user_ids = interaction_matrix['user_id'].values
item_ids = interaction_matrix['item_id'].values
strengths = interaction_matrix['strength'].values

# разбиваем на train и test по взаимодействиям
X_train, X_test = train_test_split(
    list(zip(user_ids, item_ids, strengths)),
    test_size=0.2,
    random_state=SEED
)

# создаём пустые разреженные матрицы в формате LIL (удобно для заполнения)
train_matrix = lil_matrix(R.shape)
test_matrix = lil_matrix(R.shape)

# заполняем матрицы
for u, i, s in X_train:
    train_matrix[u, i] = s
for u, i, s in X_test:
    test_matrix[u, i] = s

# переводим обратно в CSR формат
train_matrix = train_matrix.tocsr()
test_matrix = test_matrix.tocsr()
```

## Обучение ALS (Alternating Least Squares)

 — метод factorization:

`R ≈ U × V^T`

где U — матрица пользователей (user factors) [n_users × factors]

V — матрица товаров (item factors) [n_items × factors]

Результат — приближённое восстановление предпочтений и возможность рекомендовать новые товары.

```python
# ALS
als_model = implicit.als.AlternatingLeastSquares(
    factors=50, regularization=0.1, iterations=30)

start_time = time.time()
als_model.fit(train_matrix)
als_time = time.time() - start_time

als_recommendations = als_model.recommend_all(train_matrix)
```

```python
als_recommendations
```

```python
# als_recommendations: (n_users × N) — список из N item_id, рекомендованных пользователю.
df_recs = pd.DataFrame(als_recommendations)
df_recs.columns = [f'top{i+1}' for i in range(df_recs.shape[1])]
df_recs['user_id'] = df_recs.index

df_recs = df_recs[['user_id'] + [f'top{i+1}' for i in range(df_recs.shape[1] - 1)]]
```

## Обучение BPR (Bayesian Personalized Ranking)

— метод обучения рекомендаций по неявной обратной связи (implicit feedback), оптимизирующий ранжирование товаров, а не приближение рейтингов как ALS.

BPR обучается так, чтобы u предпочитал i перед j:

`score(u, i) > score(u, j)`

((пользователь u, товар i, с которым он взаимодействовал (например, просмотрел, кликнул), товар j, с которым он не взаимодействовал))

```python
# BPR
bpr_model = implicit.bpr.BayesianPersonalizedRanking(
    factors=50, regularization=0.1, iterations=30)

start_time = time.time()
bpr_model.fit(train_matrix)
bpr_time = time.time() - start_time

bpr_recommendations = bpr_model.recommend_all(train_matrix)
```

## Оценка моделей

```python
from scipy.sparse import csr_matrix

train_matrix = train_matrix.tocsr()
test_matrix = test_matrix.tocsr()
```

```python
import numpy as np

def precision_recall_at_k(true_matrix, pred_matrix, k=10):
    true_items = true_matrix.tolil().rows
    pred_items = np.argsort(-pred_matrix, axis=1)[:, :k]

    precisions = []
    recalls = []

    for user_id in range(len(true_items)):
        true_set = set(true_items[user_id])
        pred_set = set(pred_items[user_id])

        if len(true_set) > 0:
            precision = len(true_set & pred_set) / k
            recall = len(true_set & pred_set) / len(true_set)
            precisions.append(precision)
            recalls.append(recall)

    return np.mean(precisions), np.mean(recalls)

# предсказанные рейтинги всех пользователей ко всем товарам
als_pred_matrix = als_model.user_factors @ als_model.item_factors.T
bpr_pred_matrix = bpr_model.user_factors @ bpr_model.item_factors.T

als_precision, als_recall = precision_recall_at_k(test_matrix, als_pred_matrix, k=10)
bpr_precision, bpr_recall = precision_recall_at_k(test_matrix, bpr_pred_matrix, k=10)

print(f"ALS Precision@10: {als_precision:.2f}")
print(f"ALS Recall@10: {als_recall:.2f}")
print()
print(f"BPR Precision@10: {bpr_precision:.2f}")
print(f"BPR Recall@10: {bpr_recall:.2f}")
```
